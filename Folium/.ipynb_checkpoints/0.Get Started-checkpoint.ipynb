```python
# print out a Dublin map
import folium as fm
dub_map= fm.Map(location=[53.3498, -6.2603],
               titles='Dublin Map')
dub_map
```

```python
# Add 2 nice coffee marker
tooltip = 'Click me!'

fm.Marker([53.3480, -6.2478], popup='<i>CoffeeAngle 5 Stars</i>', tooltip=tooltip, icon=fm.Icon(color='green',icon='coffee',prefix='fa')).add_to(dub_map)
fm.Marker([53.3426, -6.2632], popup='<b>Kaph 5 Starts</b>', tooltip=tooltip, icon=fm.Icon(color='gray',icon='coffee',prefix='fa')).add_to(dub_map)

dub_map
```
