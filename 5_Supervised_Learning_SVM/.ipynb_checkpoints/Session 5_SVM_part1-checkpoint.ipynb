In this notebook, we will cover a classic classification problem - To identify what type of iris this given flower is

To get the dataset, you could visit here: https://www.kaggle.com/arshid/iris-flower-dataset

#### Background of this problem

The Iris flower data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

This dataset became a typical test case for many statistical classification techniques in machine learning such as support vector machines

```python
# Step 1 - import data

import pandas as pd

df = pd.read_csv('~/Machine_Learning_101/5_Supervised_Learning_SVM/IRIS.csv')
df.info()
```

```python
# Step 2 - Find my dependent variable

y=df['species']
y.unique()
```

#### If we only consider 2 different output - labels - let's say 'Iris-setosa' and 'Iris-versicolor' and 2 different inputs - sepal_length, sepal_width as our starting point

```python
df_small=df[['sepal_length','sepal_width','species']][df['species'] != 'Iris-virginica']
X=df_small[['sepal_length','sepal_width']]
Y=df_small['species']

y_mapdict = {'Iris-setosa': 0, 'Iris-versicolor': 1}
Y = Y.map(y_mapdict)

from sklearn.svm import SVC

cls = SVC(kernel='linear', C=1.5) 
cls.fit(X, Y)

w = cls.coef_  
b = cls.intercept_
sv_idx = cls.support_ 
n_Support_vector = cls.n_support_

import matplotlib.pyplot as plt
import numpy as np
# from mpl_toolkits.mplot3d import Axes3D
from mlxtend.plotting import plot_decision_regions

%matplotlib inline

plot_decision_regions(X.values, 
                      Y.values,
                      clf=cls, 
                      legend=2)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris')
plt.show()

print (w)
print(b)
print(sv_idx)
print(n_Support_vector)

```

```python
iris_mapdict={'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
y=y.map(iris_mapdict)
```

```python
# Step 3 - build model

# split data into training and testing data

from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(df.loc[:,:'petal_width'],y,train_size=0.8)

from sklearn.svm import SVC
clf = SVC(gamma='auto')
clf.fit(train_x,train_y)
```

```python
print ('Number of Support Vectors of each class: ', clf.n_support_)
print ('Should return fit status as 0 which means correctly fitted: ', clf.fit_status_)
print ('Number of Support Vectors: ', clf.n_support_)
print ('Display Support Vectors: ', clf.support_vectors_)
```
