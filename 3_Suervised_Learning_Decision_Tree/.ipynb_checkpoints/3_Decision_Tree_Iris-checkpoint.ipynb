```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
iris
```

```python
X = iris.data[:,2:] 
Y = iris.target
print(X[:5])
print(Y[:5])
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2) 
tree_clf.fit(X,Y)
```

```python
from sklearn.tree import export_graphviz

dot_file = export_graphviz (tree_clf,
                out_file = ("iris_tree.dot"),
                feature_names = iris.feature_names[2:],
                class_names = iris.target_names,
                rounded = True,
                filled = True)
```

```python
!dot -Tpng iris_tree.dot -o iris_tree.png
```

output


Notes:

DecisionTreeClassifier(max_depth=2)  - number of max_depth means the number of layers
root node == depth 0
a node's samples attribute counts how many traning instances it applies to.
gini ==0 means this node is "pure" - all training instances it applies to belong to the same class

Scikit-learn uses the CART algorithm, which produces only binary trees: nonleaf nodes always have 2 children (yes/no)
ID3 algorithm can produce DT with nodes that have more than 2 children

DT model is eay to interpret so this model is often called White Box Model


```python
tree_clf.predict_proba([[5,1.5]])
```

```python
tree_clf.predict([[5,1.5]])
```

Scikit-learn DT by default use Gini impurity but you could set it as entropy.  Most of the time it does not make a big difference: they lead to similar trees.

Gini impurti is slightly faster to compute. However, Gini impurity tends to isolate hte most frequent class in its own branch of the tree, while entropy tends to produce slightly more balance trees.

```python
tree_clf_entropy = DecisionTreeClassifier(criterion="entropy",max_depth=2) 
tree_clf_entropy.fit(X,Y)
from sklearn.tree import export_graphviz

dot_file = export_graphviz (tree_clf_entropy,
                out_file = ("iris_tree_entropy.dot"),
                feature_names = iris.feature_names[2:],
                class_names = iris.target_names,
                rounded = True,
                filled = True)
```

```python
!dot -Tpng iris_tree_entropy.dot -o iris_tree_entropy.png
```
