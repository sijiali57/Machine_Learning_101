### Step 1 - import data and describe dataset

Data source: https://www.kaggle.com/ronitf/heart-disease-uci

*Context*

This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them. In particular, the Cleveland database is the only one that has been used by ML researchers to this date. The "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4.

*Attribute Information: *

> 1. age 
> 2. sex 
> 3. chest pain type (4 values) 
> 4. resting blood pressure 
> 5. serum cholestoral in mg/dl 
> 6. fasting blood sugar > 120 mg/dl
> 7. resting electrocardiographic results (values 0,1,2)
> 8. maximum heart rate achieved 
> 9. exercise induced angina 
> 10. oldpeak = ST depression induced by exercise relative to rest 
> 11. the slope of the peak exercise ST segment 
> 12. number of major vessels (0-3) colored by flourosopy 
> 13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

### Step 2 - Get subset data

```python
# Get subset for 1st Model
inputData_1 = df[['age','trestbps','target']]
inputData_1[:10]
```

```python
# Print out the data plot
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

scatter=ax.scatter(df['age'],df['trestbps'],marker ='o', c=df['target'], s=df['cp']*50)  

legend = ax.legend(*scatter.legend_elements(), loc="upper right", title ="Target")

ax.add_artist(legend)

plt.xlabel("Age")
plt.ylabel('Resting blood pressure')
plt.title('Heart Disease')
plt.show()
```

### Step 3 - Build model

```python
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier as knn

knn_model_2input = knn(n_neighbors =3)
knn_model_2input.fit(df[['age','trestbps']],df['target'])

model_graph = knn_model_2input.kneighbors_graph(df[['age','trestbps']])
model_graph.toarray()
```

### Step 4 - Predict 

```python

new_input1=pd.DataFrame({'age':[30], 
              'trestbps':[110]})
print(knn_model_2input.predict(new_input1))
```

```python
new_input1=pd.DataFrame({'age':[60], 
              'trestbps':[138]})
print(knn_model_2input.predict(new_input1))
```

## Model 2 - Full data

```python
# Set K=3

X=df.loc[:,:'thal']
Y=df['target']

# Split data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.8)

# Build model
from sklearn.neighbors import KNeighborsClassifier as knn
knn_model_full = knn(n_neighbors =3)
knn_model_full.fit(train_x,train_y)


# Accuracy
from sklearn.metrics import accuracy_score
knn_model_full.score(test_x,test_y)

```

### How to define K value

```python
%%time
# Set K from 1 to 5 and test which one is best

# Split data
# from sklearn.model_selection import train_test_split
# train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.8)

# Build model
# from sklearn.neighbors import KNeighborsClassifier as knn
# from sklearn.metrics import accuracy_score

for i in range(1, 20):
    knn_model_k=knn(n_neighbors= i)
    knn_model_k.fit(train_x,train_y)
    print ('K value is: ', i, ' Model Accuracy Score is: ', knn_model_k.score(test_x,test_y))
    print ('---------------------------')


```

### How to define search algorithm 

according to above results, we know when k = 8 the accuracy score is the highest one

since KNN simply remember all of its training data (possibly transformed into a fast indexing structure such as a Ball Tree or KD Tree, now next step is to find which indexing method run fastest

```python
%%time
knn_model_k=knn(algorithm='brute',n_neighbors=8)
knn_model_k.fit(train_x,train_y)
print ('K value is: ', i, ' Model Accuracy Score is: ', knn_model_k.score(test_x,test_y))
print ('---------------------------')
```

```python
%%time
knn_model_k=knn(algorithm='kd_tree',n_neighbors=8)
knn_model_k.fit(train_x,train_y)
print (' Model Accuracy Score is: ', knn_model_k.score(test_x,test_y))
print ('---------------------------')
```

```python
%%time
knn_model_k=knn(algorithm='ball_tree',n_neighbors=8)
knn_model_k.fit(train_x,train_y)
print ('K value is: ', i, ' Model Accuracy Score is: ', knn_model_k.score(test_x,test_y))
print ('---------------------------')
```

With above results, for this example, kd_tree indexing is the best

#### Conclusion

When k=8, the model has the best fit 67% - which means this model 67/100 observations correctly.

This accuracy rate is still not good enough, and now let's increase the training dataset to validate the data



```python
# Split data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.95)

# Build model
from sklearn.neighbors import KNeighborsClassifier as knn
knn_model_full = knn(n_neighbors =5,algorithm='kd_tree')
knn_model_full.fit(train_x,train_y)


# Accuracy
from sklearn.metrics import accuracy_score
knn_model_full.score(test_x,test_y)
```

After increase the training dataset, the accuracy score improved a lot. This means more data would certainly help. 

However, 75% is still not good engough. here score in sklearn returns the mean accuracy on the given test data and labels.

KNN might not be the optimal machine learning method for this problem.

```python
%%time
# Recap previous Decision Tree model
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(train_x,train_y)
dt_model.score(test_x,test_y)
```

Decision Tree returns better perfomance than KNN in this case but 81% accuracy is still not good enough
